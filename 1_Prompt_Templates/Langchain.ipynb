## API Key Setup

```python
from langchain_openai import ChatOpenAI
import os
from dotenv import load_dotenv

load_dotenv()

llm = ChatOpenAI(
    model="nvidia/nemotron-3-super-120b-a12b:free",
    openai_api_key=os.getenv("OPENROUTER_API_KEY"),
    openai_api_base="https://openrouter.ai/api/v1"
)
```

## Prompt Engineering
* Prompt Engineering is about guiding the model's behaviour without changing it's underlying capabilities. By providing instructions, examples, and context, prompts shape how the model responds to input.
* Prompt Engineering allows you to modify the model's behaviour.

### Two types in Prompt
* 1. System Prompt  :-------> Means that you are givving certain instructions to LLMs to follow
* 2. User Prompt :-------> It is a normal user query

## Basic Prompt Template

```python
from langchain_core.prompts import PromptTemplate

template = """
Explain {topic} in simple terms.
"""

prompt = PromptTemplate(
    input_variables=["topic"],
    template=template
)

final_prompt = prompt.format(topic="Neural Networks")

response = llm.invoke(final_prompt)

print(response.content)
```

## Multiple variables in Prompt Template

```python
template = """
Explain {topic} for a {audience}.
Give 3 examples.
"""

prompt1 = PromptTemplate(
    input_variables=["topic", "audience"],
    template=template
)

final_prompt1 = prompt1.format(
    topic="Machine Learning",
    audience="high school students"
)

response = llm.invoke(final_prompt1)

print(response.content)
```

## ChatPrompt Template

```python
from langchain_core.prompts import ChatPromptTemplate

prompt = ChatPromptTemplate.from_template(
    "Explain {topic} in simple terms."
)

chain = prompt | llm

response = chain.invoke({"topic": "LLMs"})

print(response.content)
```

```python
prompt = ChatPromptTemplate.from_messages([
    ("system", "You are an expert AI teacher."),
    ("human", "Explain {topic}")
])

chain = prompt | llm

response = chain.invoke({"topic": "LangChain"})

print(response.content)
```

## Few-Shot Prompt Template

```python
from langchain_core.prompts import FewShotPromptTemplate

examples = [
    {"question": "2+2", "answer": "4"},
    {"question": "3+5", "answer": "8"},
]

example_template = """
Question: {question}
Answer: {answer}
"""

example_prompt = PromptTemplate(
    input_variables=["question", "answer"],
    template=example_template
)

few_shot_prompt = FewShotPromptTemplate(
    examples=examples,
    example_prompt=example_prompt,
    prefix="Solve the following math problems:",
    suffix="Question: {input}",
    input_variables=["input"]
)

chain = few_shot_prompt | llm

# Ask a new question
response = chain.invoke({"input": "6+7"})

print(response.content)
```
