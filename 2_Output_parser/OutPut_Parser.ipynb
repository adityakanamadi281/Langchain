```python
from langchain_openai import ChatOpenAI
import os
from dotenv import load_dotenv

load_dotenv()

llm = ChatOpenAI(
    model="nvidia/nemotron-3-super-120b-a12b:free",
    openai_api_key=os.getenv("OPENROUTER_API_KEY"),
    openai_api_base="https://openrouter.ai/api/v1",
    temperature=0
)
```

```python
response1 = llm.invoke("Give me a list of countries and their income source.")
print(response1.content)
```

## StrOutputParser
* To extract plain text

```python
from langchain_core.output_parsers import StrOutputParser
from langchain_core.prompts import ChatPromptTemplate

prompt = ChatPromptTemplate.from_template(
    "Explain {topic} in one sentence."
)

parser = StrOutputParser()

chain = prompt | llm | parser

result = chain.invoke({"topic": "Investment Banking"})

print(result)
```

## CommaSeparatedListOutputParser

```python
from langchain_core.output_parsers import CommaSeparatedListOutputParser
from langchain_core.prompts import PromptTemplate

parser = CommaSeparatedListOutputParser()

prompt = PromptTemplate(
    template="List 5 AI startups separated by commas.\n{format_instructions}",
    input_variables=[],
    partial_variables={"format_instructions": parser.get_format_instructions()},
)

chain = prompt | llm | parser

result = chain.invoke({})

print(result)
```

## JSONOutputParser
* To extract Structured JSON-like output

```python
from langchain_core.output_parsers import JsonOutputParser
from langchain_core.prompts import PromptTemplate

parser1 = JsonOutputParser()

prompt1 = PromptTemplate(
    template="Return a JSON with keys 'name' and 'description' about an AI startup.",
    input_variables=[]
)

chain = prompt1 | llm | parser1

result = chain.invoke({})

print(result)
```

## PydanticOutputParser


```python
from langchain_core.output_parsers import PydanticOutputParser
from pydantic import BaseModel, Field

class StartupIdea(BaseModel):
    name: str = Field(description="startup name")
    description: str = Field(description="startup description")


parser2 = PydanticOutputParser(pydantic_object=StartupIdea)

prompt2 = PromptTemplate(
    template="Generate an AI startup idea.\n{format_instructions}",
    input_variables=[],
    partial_variables={"format_instructions": parser2.get_format_instructions()},
)

chain2 = prompt2 | llm | parser2

result2 = chain2.invoke({})

print(result2)
print(result2.name)
print(result2.description)
```
